```python
from transformers import pipeline
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain_community.llms import HuggingFacePipeline
```

```python
generate = pipeline(
    model="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=500,
    do_sample=True,
    temperature=0.8,
    top_p=0.9
    #repitition_penalty=1.15
)

llm = HuggingFacePipeline(pipeline=generate)
```

```python
ad_prompt = PromptTemplate(
    input_variables=["bike_type", "features", "discount", "theme"],
    template="""
    You are an advertisement assistant for BikeEase.

    Generate a creative BikeEase advertisement for a bike rental.

    Example (style + format):
    Headline: Ride farther, feel freer
    Body: Take on your day with a smooth Hybrid bike built for comfort and quick trips. Book fast and explore more with BikeEase.

    Utilize these inputs for the new advertisement and make sure all different inputs make it into the advert in a way :

    Bike type: {bike_type}
    Features: {features}
    Theme: {theme} (marketing of the ad)
    Discount: {discount}

    Return ONLY:
    Headline:
    Body: (Write exactly THREE full sentences. Each sentence should add new information. Do not stop early.)
"""
)
```

```python
ad_chain = LLMChain(llm=llm, prompt=ad_prompt)
```

```python
#Bike Ad user input

def bikeAd():
  bike_type = input("Bike Type: ").strip()
  features = input("Features: ").strip()
  discount = input("Discount: ").strip()
  theme = input("Theme: ").strip()
  return bike_type, features, discount, theme

def generateAd():
  bike_type, features, discount, theme = bikeAd()

  if discount.lower() in ["none", "no discount", ""]:
    discount = "None"

  ad = ad_chain.run(
      bike_type=bike_type,
      features=features,
      discount=discount,
      theme=theme
  )

  print("\nGenerated Advertisement\n")
  print(ad)


```

```python
generateAd()
```
