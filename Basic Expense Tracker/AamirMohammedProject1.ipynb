```python
# Expense Tracker
import csv
import pandas as pd
from datetime import datetime
```

```python
dataFile = "expenses.csv"
col = ["date", "category", "amount", "description"]

df = pd.DataFrame(columns=col)
expense_budget = None
```

```python
#1 Add an expense:

def add_expense(date, cat, amnt, des):
    # User Date
    try:
        dt = datetime.strptime(str(date).strip(), "%Y-%m-%d")
    except ValueError:
        raise ValueError("The Date must be in format: YYYY-MM-DD (e.g., 2024-09-18).")
    fullDate = dt.strftime("%Y-%m-%d")
    
    # Category
    category = str(cat).strip()
    if not category:
        raise ValueError("A Category is required")
        
    # Amount
    try:
        amount = float(amnt)
        if amount < 0:
            print("Amount must be non-negative")
    except (TypeError, ValueError):
        raise ValueError("Amount must be a number")
        
        
    # Description
    desc = str(des).strip()
    if not desc:
        raise ValueError("A Description is required")
        
    
    expenses = {"date": fullDate,
                "category": category.title(),
                "amount": round(amount, 2),
                "description": desc}
    
    return expenses
```

```python
# User Input for Dictionary

def user_expense():
    while True:
        try:
            string_date = input("Please input Date of expense(YYYY-MM-DD): ").strip()
            cat = input("Please input Category (Food, Travel, etc): ").strip()
            amt = input("Please input Amount spent (e.g., 20.50): ").strip()
            desc = input("Please input a Description of the Expense: ").strip()
            
            expenses = add_expense(string_date, cat, amt, desc)
            df.loc[len(df)] = expenses
            
            print("Expense Added to File!", expenses)
            break
        except:
            print("Incorrect input, please try again!")
```

```python
# Viewing Expenses

def view_expenses():
    if df.empty:
        print("No Expenses Recorded in file yet")
        return
    
    print(f"{'Date':<12} {'Category':<12} {'Amount':>10} Description")
    print("-"*50)
    
    
    for _, row in df.iterrows():
        if all([row.get("date"), row.get("category"), row.get("amount"), row.get("description")]):
            print(f"{row['date']:<12} {row['category']:<12} {float(row['amount']):>10.2f} {row['description']}")
            
    
```

```python
# Set and Track Monthly Budget

def set_budget():
    global expense_budget
    while True:
        try:
            expense_budget = float(input("Please Enter your Monthy Budget: "))
            if expense_budget < 0:
                print("The Budget must not be negative!")
                continue
                
            print("The Budget is set to: ", round(expense_budget, 2))
            break
        except ValueError:
            print("Please enter a valid Budget Number!")
            


def track_budget():
    if df.empty:
        print("No Expenses Recorded in file yet")
        return
    
    if expense_budget is None:
        print("A Budget has not been set yet!")
        return
    
    tot = df['amount'].sum()
    rem = expense_budget - tot
    
    print("Monthly Budget: ", round(expense_budget, 2))
    print("Total spent:    ", round(tot,2 ))
    
    if rem < 0:
        print("You have exceeded you budget by: ", -rem)
    else:
        print("You have ", rem, " left for the month!")
    
            
```

```python
# Saving and Loading the FILE

def save_expenses():
    if df.empty:
        print("No Expenses Recorded in file to save")
        return
    df.to_csv(dataFile, index=False)
    print(f"Saved {len(df)} expense(s) to {dataFile}")
```

```python
def load_expenses():
    global df
    try:
        df = pd.read_csv(dataFile)
        print(f"Loading {len(df)} expense(s) from {dataFile}")
    except FileNotFoundError:
        print("No File Found! let's try again.")
        df = pd.DataFrame(columns=col)        
              
```

```python
#Interactive User Menu
def menu():
    while True:
        print("\n === PERSONAL EXPENSE TRACKER ===")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Track Budget")
        print("4. Save Expenses")
        print("5. Exit Menu")
        
        user_choice = input("Please Enter your choice (1-5): ")
        
        if user_choice == "1":
            user_expense()
        elif user_choice == "2":
            view_expenses()
        elif user_choice == "3":
            set_budget()
            track_budget()
        elif user_choice == "4":
            save_expenses()
        elif user_choice == "5":
            save_expenses()
            print("Expenses Are Saved to File, exiting program")
            break
        else:
            print("Invalid option, please try again!")
```

```python
load_expenses()

menu()
```
